```python
img = np.array([np.ones((41,))*10-np.abs(np.arange(-20, 21))])*np.array([np.ones((41,))*10-np.abs(np.arange(-20, 21))]).T

fig=plt.figure(figsize = (4.5,6), dpi=200)
gss = gridspec.GridSpec(4, 3, figure=fig)
gss.update(hspace=0.1, wspace=0)

ax_top = fig.add_subplot(gss[0,:])
for i, gs in enumerate(gss):
    if not gs.is_first_row():
        ax = fig.add_subplot(gs) 
        ax.imshow(img) 
        ax.set_axis_off()

locuses = [200, 500, 1400]
chrom = Chromosome(1800, 'chr1', centromere=700, locuses=locuses)

h=9.5
original_grid = np.linspace(0, 100, len(locuses)+2)[1:-1]
grid_shift = np.array([-h]*(len(locuses)//2) + [0]*(len(locuses)%2) + [h]*(len(locuses)//2))
grid = original_grid + grid_shift
chrom.plot_chromosome(ax=ax_top, locuses_x_grid=grid)
 

plt.show()
```
